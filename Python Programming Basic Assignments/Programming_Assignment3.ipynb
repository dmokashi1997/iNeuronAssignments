```python
## WAP to Check if a Number is Positive, Negative or Zero
def checkNumber(number):
    if number==0:
        print("Number is zero")
    elif number>0:
        print("Given number {0} is positive".format(number))
    else:
        print("Given number {0} is negative".format(number))

checkNumber(1)
checkNumber(0)
checkNumber(-98)
checkNumber(1000000)
```

```python
##WAP to Check if a Number is Odd or Even

def oddEven(number):
    if(number%2==0):
        return "Number is Even"
    else:
        return "Number is Odd"
    
print(oddEven(3))
print(oddEven(1200))
```

```python
## WAP to Check Leap Year

def checkLeapYear(year):
    if (year % 400 == 0) and (year % 100 == 0):
        print("{0} is a leap year".format(year))
    elif (year % 4 ==0) and (year % 100 != 0):
        print("{0} is a leap year".format(year))
    else:
         print("{0} is not a leap year".format(year))
            
checkLeapYear(2022)
checkLeapYear(2020)
```

```python
## WAP to Check Prime Number?

def checkPrime(number):
    flag=False
    if number==1:
        flag=False
    else:
        for i in range(2,number):
            if(number%i==0):
                flag=False
                break
            else:
                flag=True
                continue
     
    return flag
    
        
flag=checkPrime(9)
if flag==True:
    print("Given number is a prime number")
else:
    print("Given number is not prime")
```

```python
## WAP to print all Prime Numbers in an Interval of 1-10000

def primeInRange():
    for i in range(0,100001):
        flag=checkPrime(i)
        if(flag==True):
            print("{0} is a prime number".format(i))
primeInRange()
```
